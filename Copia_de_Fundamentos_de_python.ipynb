Ejercicio 2: Calcular el valor a pagar según los metros consumidos de agua potable.

```python
def calcular_valor_agua(metros_consumidos):
    if metros_consumidos <= 18:
        return 6
    elif 19 <= metros_consumidos <= 28:
        exceso = metros_consumidos - 18
        return 6 + (exceso * 0.45)
    else:
        exceso = metros_consumidos - 28
        return 6 + (10 * 0.45) + (exceso * 0.65)


metros_consumidos = int(input("Ingrese el número de los metros consumidos: "))
valor_a_pagar = calcular_valor_agua(metros_consumidos)
print(f"Valor a pagar: ${valor_a_pagar:.2f}")
```

```python
# Funciones de conversión para Longitud
def convertir_longitud(valor, unidad_origen, unidad_destino):
    conversiones = {
        'm': 1,
        'cm': 100,
        'mm': 1000,
        'km': 0.001,
        'in': 39.3701,
        'ft': 3.28084,
        'yd': 1.09361,
        'mi': 0.000621371,
        'nm': 1000000000,
        'nmi': 0.000539957
    }
    metros = valor / conversiones[unidad_origen]
    return metros * conversiones[unidad_destino]

# Funciones de conversión para Masa
def convertir_masa(valor, unidad_origen, unidad_destino):
    conversiones = {
        'g': 1,
        'kg': 0.001,
        'mg': 1000,
        't': 0.000001,
        'lb': 0.00220462,
        'oz': 0.035274,
        'q': 0.00001,
        'µg': 1000000,
        'st': 0.000157473,
        'dal': 0.001
    }
    gramos = valor / conversiones[unidad_origen]
    return gramos * conversiones[unidad_destino]

# Funciones de conversión para Almacenamiento
def convertir_almacenamiento(valor, unidad_origen, unidad_destino):
    conversiones = {
        'B': 1,
        'KB': 0.001,
        'MB': 1e-6,
        'GB': 1e-9,
        'TB': 1e-12,
        'PB': 1e-15,
        'EB': 1e-18,
        'ZB': 1e-21,
        'YB': 1e-24,
        'nibble': 0.5
    }
    bytes = valor / conversiones[unidad_origen]
    return bytes * conversiones[unidad_destino]

# Funciones de conversión para Tiempo
def convertir_tiempo(valor, unidad_origen, unidad_destino):
    conversiones = {
        's': 1,
        'min': 1/60,
        'h': 1/3600,
        'd': 1/86400,
        'wk': 1/604800,
        'mo': 1/2592000,
        'yr': 1/31536000,
        'ms': 1000,
        'µs': 1e6,
        'ns': 1e9
    }
    segundos = valor / conversiones[unidad_origen]
    return segundos * conversiones[unidad_destino]

# Funciones de conversión para Volumen
def convertir_volumen(valor, unidad_origen, unidad_destino):
    conversiones = {
        'L': 1,
        'mL': 1000,
        'cm³': 1000,
        'm³': 0.001,
        'gal': 0.264172,
        'pt': 2.11338,
        'fl oz': 33.814,
        'tsp': 202.884,
        'tbsp': 67.628,
        'bbl': 0.00628981
    }
    litros = valor / conversiones[unidad_origen]
    return litros * conversiones[unidad_destino]

# Funciones de conversión para Área
def convertir_area(valor, unidad_origen, unidad_destino):
    conversiones = {
        'm²': 1,
        'cm²': 10000,
        'km²': 0.000001,
        'ha': 0.0001,
        'ac': 0.000247105,
        'ft²': 10.7639,
        'yd²': 1.19599,
        'in²': 1550.0031,
        'mi²': 3.861e-7,
        'a': 0.01
    }
    metros_cuadrados = valor / conversiones[unidad_origen]
    return metros_cuadrados * conversiones[unidad_destino]

# Funciones de conversión para Manzanas
def convertir_manzanas(valor, unidad_origen, unidad_destino):
    conversiones = {
        'manzana': 0.0064,
        'cuerda': 0.0069,
        'ha': 0.01,
        'ac': 0.0247105,
        'km²': 0.00001,
        'm²': 6400,
        'ft²': 68814.4,
        'yd²': 7680,
        'hato': 0.0064,
        'tarea': 0.01
    }
    hectareas = valor / conversiones[unidad_origen]
    return hectareas * conversiones[unidad_destino]

# Funciones de conversión para Tareas
def convertir_tareas(valor, unidad_origen, unidad_destino):
    conversiones = {
        'min': 1,
        'h': 1/60,
        'd': 1/1440,
        'wk': 1/10080,
        'mo': 1/43200,
        'yr': 1/525600,
        's': 60,
        'turno': 1/480,
        'reunión': 1/60,
        'intervalo': 1/60
    }
    minutos = valor / conversiones[unidad_origen]
    return minutos * conversiones[unidad_destino]

    # Funciones de conversión para Moneda
def convertir_moneda(valor, moneda_origen, moneda_destino):
    conversiones = {
        'USD': 1,          # Dólar estadounidense
        'EUR': 0.85,       # Euro
        'JPY': 110,        # Yen japonés
        'GBP': 0.75,       # Libra esterlina
        'AUD': 1.35,       # Dólar australiano
        'CAD': 1.25,       # Dólar canadiense
        'CHF': 0.92,       # Franco suizo
        'CNY': 6.45,       # Yuan chino
        'MXN': 20.0,       # Peso mexicano
        'BRL': 5.25,        # Real brasileño
        'SVC': 8.73        # Colón salvadoreño
    }
    dolares = valor / conversiones[moneda_origen]
    return dolares * conversiones[moneda_destino]

# Menú de selección
def menu():
    print("Seleccione la categoría para la conversión:")
    print("1. Longitud")
    print("2. Masa")
    print("3. Almacenamiento")
    print("4. Tiempo")
    print("5. Volumen")
    print("6. Área")
    print("7. Manzanas")
    print("8. Tareas")
    print("9. Moneda")
    print("10. Salir")

    seleccion = int(input("Ingrese el número de la categoría: "))

    if seleccion == 10:
        return

    if seleccion == 1:
        categoria = 'Longitud'
        conversion_func = convertir_longitud
        unidades = ['m', 'cm', 'mm', 'km', 'in', 'ft', 'yd', 'mi', 'nm', 'nmi']
    elif seleccion == 2:
        categoria = 'Masa'
        conversion_func = convertir_masa
        unidades = ['g', 'kg', 'mg', 't', 'lb', 'oz', 'q', 'µg', 'st', 'dal']
    elif seleccion == 3:
        categoria = 'Almacenamiento'
        conversion_func = convertir_almacenamiento
        unidades = ['B', 'KB', 'MB', 'GB', 'TB', 'PB', 'EB', 'ZB', 'YB', 'nibble']
    elif seleccion == 4:
        categoria = 'Tiempo'
        conversion_func = convertir_tiempo
        unidades = ['s', 'min', 'h', 'd', 'wk', 'mo', 'yr', 'ms', 'µs', 'ns']
    elif seleccion == 5:
        categoria = 'Volumen'
        conversion_func = convertir_volumen
        unidades = ['L', 'mL', 'cm³', 'm³', 'gal', 'pt', 'fl oz', 'tsp', 'tbsp', 'bbl']
    elif seleccion == 6:
        categoria = 'Área'
        conversion_func = convertir_area
        unidades = ['m²', 'cm²', 'km²', 'ha', 'ac', 'ft²', 'yd²', 'in²', 'mi²', 'a']
    elif seleccion == 7:
        categoria = 'Manzanas'
        conversion_func = convertir_manzanas
        unidades = ['manzana', 'cuerda', 'ha', 'ac', 'km²', 'm²', 'ft²', 'yd²', 'hato', 'tarea']
    elif seleccion == 8:
        categoria = 'Tareas'
        conversion_func = convertir_tareas
        unidades = ['min', 'h', 'd', 'wk', 'mo', 'yr', 's', 'turno', 'reunión', 'intervalo']
    elif seleccion == 9:
        categoria = 'Moneda'
        conversion_func = convertir_moneda
        unidades = ['USD', 'EUR', 'JPY', 'GBP', 'AUD', 'CAD', 'CHF', 'CNY', 'MXN', 'BRL', 'SVC']
    else:
        print("Selección no válida")
        return

    print(f"\nCategoría: {categoria}")
    print("Unidades disponibles:")
    for i, unidad in enumerate(unidades):
        print(f"{i + 1}. {unidad}")

    origen = int(input("\nSeleccione la unidad de origen: ")) - 1
    destino = int(input("Seleccione la unidad de destino: ")) - 1

    valor = float(input("Ingrese el valor a convertir: "))

    resultado = conversion_func(valor, unidades[origen], unidades[destino])
    print(f"\n{valor} {unidades[origen]} = {resultado} {unidades[destino]}\n")

# Ejecutar el menú
menu()
```

Escriba un programa que determine si una palabra introducida por el usuario es palindrome.

```python
def es_palindromo(palabra):
    palabra = palabra.replace(" ", "").lower()
    return palabra == palabra[::-1]

entrada = input("Introduce una palabra para verificar si es un palíndromo: ")

if es_palindromo(entrada):
    print(f"La palabra '{entrada}' es un palíndromo.")
else:
    print(f"La palabra '{entrada}' no es un palíndromo.")
```

```python
def calcular_impuesto(monto):
    tabla_impuestos = [
        {'desde': 0.01, 'hasta': 500, 'precio_fijo': 1.5, 'adicional': 0},
        {'desde': 500.01, 'hasta': 1000, 'precio_fijo': 1.5, 'adicional': 3},
        {'desde': 1000.01, 'hasta': 2000, 'precio_fijo': 3, 'adicional': 3},
        {'desde': 2000.01, 'hasta': 3000, 'precio_fijo': 6, 'adicional': 3},
        {'desde': 3000.01, 'hasta': 6000, 'precio_fijo': 9, 'adicional': 2},
        {'desde': 8000.01, 'hasta': 18000, 'precio_fijo': 15, 'adicional': 2},
        {'desde': 18000.01, 'hasta': 30000, 'precio_fijo': 39, 'adicional': 2},
        {'desde': 30000.01, 'hasta': 60000, 'precio_fijo': 63, 'adicional': 1},
        {'desde': 60000.01, 'hasta': 100000, 'precio_fijo': 93, 'adicional': 0.8},
        {'desde': 100000.01, 'hasta': 200000, 'precio_fijo': 125, 'adicional': 0.7},
        {'desde': 20000.01, 'hasta': 400000, 'precio_fijo': 195, 'adicional': 0.6},
        {'desde': 400000.01, 'hasta': 600000, 'precio_fijo': 255, 'adicional': 0.45},
        {'desde': 600000.01, 'hasta': 800000, 'precio_fijo': 300, 'adicional': 0.4},
        {'desde': 800000.01, 'hasta': 1000000, 'precio_fijo': 340, 'adicional': 0.3},
        {'desde': 1000000.01, 'hasta': 999999999, 'precio_fijo': 490, 'adicional': 0.18},
    ]

    for rango in tabla_impuestos:
        if rango['desde'] <= monto <= rango['hasta']:
            base = monto - rango['desde']
            impuesto = base / 1000 * rango['precio_fijo'] + rango['adicional']
            return round(impuesto, 2)
    return 0
monto = float(input("Ingrese el monto económico: "))
impuesto = calcular_impuesto(monto)
print(f"El impuesto a pagar es: ${impuesto}")
```

Desarrolle un programa en Python que simule a un cajero automático.
Haga uso de las estructuras de control.

```python
def simular_cajero(cantidad):
    denominaciones = [20, 10, 5, 1, 0.25]
    cantidad_restante = cantidad
    resultado = []

    for denominacion in denominaciones:
        cantidad_billetes = int(cantidad_restante // denominacion)
        cantidad_restante = round(cantidad_restante % denominacion, 2)
        resultado.append((cantidad_billetes, denominacion))

    return resultado

cantidad = float(input("Ingrese la cantidad a retirar: "))
resultado = simular_cajero(cantidad)

print(f"Para retirar {cantidad}, se entregará:")
for cantidad_billetes, denominacion in resultado:
    if cantidad_billetes > 0:
        print(f"{cantidad_billetes} billete(s) de ${denominacion}")
```
